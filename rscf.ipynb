```python

# Function to convert values
def money_to_number(value):
    if pd.isna(value):
        return None
    value = value.replace('$', '').strip()
    multiplier = 1
    if value.endswith('K'):
        multiplier = 1_000
        value = value[:-1]
    elif value.endswith('M'):
        multiplier = 1_000_000
        value = value[:-1]
    elif value.endswith('B'):
        multiplier = 1_000_000_000
        value = value[:-1]
    try:
        return float(value) * multiplier
    except ValueError:
        return None

# Apply the function
#df['Total Assets'] = df['Total Assets'].apply(money_to_number)
df['Operating Budget'] = df['Operating Budget'].apply(money_to_number)
df['Contributions Given'] = df['Contributions Given'].apply(money_to_number)
df['Median Grant Given'] = df['Median Grant Given'].apply(money_to_number)
df['Contributions Received'] = df['Contributions Received'].apply(money_to_number)

print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import StrMethodFormatter

# Filter to under $1B
df = funder_network[funder_network['Total Assets'] < 1e9]

# Create histogram bins
bins = np.linspace(0, 1e9, 41)  # 40 bins from 0 to 1B

# Separate by 'NYC Focused'
nyc_true = df[df['nyc_focus'] == True]['Total Assets']
nyc_false = df[df['nyc_focus'] == False]['Total Assets']

# Get counts for each bin
nyc_true_counts, _ = np.histogram(nyc_true, bins=bins)
nyc_false_counts, _ = np.histogram(nyc_false, bins=bins)

# Bar centers and width
bin_centers = (bins[:-1] + bins[1:]) / 2
width = bins[1] - bins[0]

# Plot stacked bars
fig, ax = plt.subplots(figsize=(12, 9))
ax.bar(bin_centers, nyc_true_counts, width=width, bottom=nyc_false_counts, label='NYC focused',color = "Orange")
ax.bar(bin_centers, nyc_false_counts, width=width, label='Other focus',color = "Grey")


# Formatting
ax.xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
plt.title("Organizations with Total Assets under $1B")
plt.ylabel("Organization Count")
plt.xlabel("Total Assets")
plt.xticks(np.arange(0,1e9, 50e6))
plt.xticks(rotation=45)
plt.yticks(np.arange(0, 250, 25))
plt.tight_layout()
plt.xlim([0, 1e9])
plt.legend()

plt.show()
```
